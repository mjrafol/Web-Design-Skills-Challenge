```python
import pandas as pd
from decimal import Decimal
```

```python
# Read the CSV file
df = pd.read_csv("cities.csv")
```

```python
# Copy the df before making any changes (dropping records with NA)
df2 = df
df3 = df2.dropna(how = "any")

# View the new df
df3
```

```python
# Check the df datatypes
df3.dtypes
```

```python
df4 = df3
```

```python
df4["Lat"] = df3["Lat"].astype("float64").round(2)
df4["Lng"] = df3["Lng"].astype("float64").round(2)
df4["Date"]= df3["Date"].apply(pd.to_datetime)
# df3["Date"]=Decimal(df3["Date"])
# df3['Date'] = pd.to_datetime(df3['Date'], format='%Y%m%d%H%M%S')
# df3["Date"] = df3["Date"].astype("datetime64[ns]")
df4
```
